```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
```

```python
# URL of page to be scraped
url = 'https://newjersey.craigslist.org/search/sss?sort=rel&query=guitar'
```

```python
response=requests.get(url)
```

```python
soup=bs(response.text,'lxml')
```

```python
print(soup.prettify())
```
